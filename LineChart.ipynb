```python
import pandas as pd
import matplotlib.pyplot as plt
unrate=pd.read_csv("~/unrate.csv")
unrate["DATE"]=pd.to_datetime(unrate["DATE"])
print(unrate.head(12))
```

Line Charts

```python
plt.plot(unrate[0:12]['DATE'],unrate[0:12]['VALUE'])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment rate,1948')
plt.show()
```

To create line charts, we use the matplotlib library, which allows us to: quickly create common plots using high-level functions, extensively tweak plots, and create new kinds of plots from the ground up.<br/>

By default, matplotlib displays a coordinate grid with: the x-axis and y-axis values ranging from -0.6 to 0.6, no grid lines, and no data.

Visual reperesentations use visual objects like dots, shapes, and lines on a gird.

Plots are a category of visual representation that allows us to easily understand the representation between variables.
